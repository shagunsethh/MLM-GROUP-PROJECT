**Project Objectives | Problem Statements**


*   Unsupervised Machine Learning: Clustering

1. Segmentation of Dataset using Unsupervised Machine Learning Clustering Algorithms

Objective:
To segment the dataset into clusters based on inherent similarities using unsupervised learning algorithms.

Detailed Explanation:
Segmentation involves dividing the dataset into distinct groups (clusters) where data points within the same cluster are more similar to each other than to those in other clusters.

Unsupervised Learning is used because the dataset lacks predefined labels or categories, and clustering algorithms are designed to discover hidden patterns or groupings in the data.

Algorithms considered:

KMeans Clustering: Groups data based on the nearest mean (centroid) using an iterative approach. Works best for well-separated, spherical clusters.

Birch Clustering: Uses hierarchical tree structures to incrementally cluster large datasets. Suitable for handling noise and datasets with varying cluster sizes.

DBSCAN Clustering: Groups data points based on density. It identifies clusters of arbitrary shapes and can handle outliers effectively.

2.   Identification of Appropriate Number of Segments or Clusters

Objective:
To determine the optimal number of clusters that represents the dataset's structure accurately.

Detailed Explanation:
Clustering results depend heavily on identifying the optimal number of clusters. Too few clusters may merge distinct groups, while too many can overfit the data and fragment natural groupings.

Techniques employed:

Silhouette Score:
Measures how well each data point fits within its cluster compared to other clusters. Scores range from -1 to 1, with higher values indicating better-defined clusters.

Davies-Bouldin Score: The algorithm is run for a range of cluster counts (e.g., k=2 to k=10 for KMeans or similar algorithms).
For each value of k, the Davies-Bouldin Score is calculated. The optimal number of clusters is the one with the lowest Davies-Bouldin Score.

3. Determination of Segment or Cluster Characteristics

Objective:
To analyze and interpret the characteristics of each identified segment or cluster.

Detailed Explanation:
After clustering, it is crucial to evaluate what defines each cluster in terms of features (variables). This involves:

Descriptive Statistics:
Calculate the mean, median, standard deviation, and range for each feature within each cluster.

Feature Importance:
Identify which features contribute most to the differentiation of clusters.
Tools like feature importance rankings or PCA (Principal Component Analysis) can assist in understanding the primary drivers.

Cluster Profiling:
Create profiles for each cluster by summarizing its key characteristics (e.g., high transaction value, low shipment weight, etc.). Using statistical and visual methods such as:

Heatmaps: To show feature distributions across clusters.

Boxplots: To compare the spread of features within clusters.

Scatterplots: To visualize relationships between pairs of variables for each cluster.

```python
#Calculating Dispersion Measures
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, pearsonr, spearmanr
from scipy.stats import t

def calculate_dispersion_measures(data):
    results = []
    for col in data.columns:
        column_data = data[col].dropna()  # Handle NaN values
        min_val = column_data.min()
        max_val = column_data.max()
        range_val = max_val - min_val
        std_dev = column_data.std()
        skewness = skew(column_data)
        kurt = kurtosis(column_data)

        results.append({
            "Variable": col,
            "Range": range_val,
            "Standard Deviation": std_dev,
            "Skewness": skewness,
            "Kurtosis": kurt
        })
    return pd.DataFrame(results)

# Function to calculate correlation (Pearson and Spearman)
def calculate_correlations(data):
    correlation_results = []
    columns = data.columns
    for i in range(len(columns)):
        for j in range(i+1, len(columns)):
            col1 = columns[i]
            col2 = columns[j]
            pearson_corr, _ = pearsonr(data[col1], data[col2])
            spearman_corr, _ = spearmanr(data[col1], data[col2])
            correlation_results.append({
                "Variable Pair": f"{col1} & {col2}",
                "Pearson Correlation": pearson_corr,
                "Spearman Correlation": spearman_corr
            })
    return pd.DataFrame(correlation_results)

dispersion_measures = calculate_dispersion_measures(ss42_non_cat[ss42_numerical_cols]) # Changed here
correlations = calculate_correlations(ss42_non_cat[ss42_numerical_cols]) # Changed here

print("Measures of Dispersion:")
print(dispersion_measures)

print("\nCorrelations:")
print(correlations)
```

```python
#calculate composite measures for non-cat variables

# Function to calculate composite measures
def calculate_composite_measures(data):
    results = []
    for col in data.columns:
        column_data = data[col].dropna()  # Handle NaN values
        mean_val = column_data.mean()
        std_dev = column_data.std()
        coefficient_variation = (std_dev / mean_val) * 100  # In percentage

        # Confidence Interval (95% CI)
        n = len(column_data)
        se = std_dev / np.sqrt(n)
        confidence_level = 0.95
        t_value = t.ppf((1 + confidence_level) / 2, n - 1)
        margin_of_error = t_value * se
        lower_bound = mean_val - margin_of_error
        upper_bound = mean_val + margin_of_error

        results.append({
            "Variable": col,
            "Coefficient of Variation (%)": coefficient_variation,
            "Confidence Interval": (lower_bound, upper_bound)
        })
    return pd.DataFrame(results)

composite_measures = calculate_composite_measures(ss42_non_cat[ss42_numerical_cols]) # Changed here

print("\nComposite Measures:")
print(composite_measures)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder

# Function to perform Chi-Square Test of Homogeneity
def chi_square_test(data, col1, col2):
    """
    Perform a Chi-Square Test of Homogeneity between two categorical variables.

    Parameters:
    - data: DataFrame containing the categorical variables.
    - col1: First categorical variable.
    - col2: Second categorical variable.

    Returns:
    - A dictionary containing the test statistic, p-value, degrees of freedom, and expected frequencies.
    """
    # Create a contingency table
    contingency_table = pd.crosstab(data[col1], data[col2])

    # Perform the Chi-Square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Create a DataFrame for expected frequencies with labels
    expected_df = pd.DataFrame(expected,
                               index=contingency_table.index,
                               columns=contingency_table.columns)

    return {
        "Chi-Square Statistic": chi2,
        "P-Value": p,
        "Degrees of Freedom": dof,
        "Expected Frequencies": expected_df  # Return the DataFrame
    }

# Assuming ss42_ss is your DataFrame
ss42_categorical_cols_encoded = ss42_ss[['Country', 'Category']].copy()
ss42_categorical_cols_encoded = ss42_categorical_cols_encoded.reset_index(drop=True)

# Encode categorical columns
label_encoder = LabelEncoder()
for col in ['Country', 'Category']:
    ss42_categorical_cols_encoded[col] = label_encoder.fit_transform(ss42_categorical_cols_encoded[col])

# Perform Chi-Square test
results = chi_square_test(ss42_categorical_cols_encoded, 'Country', 'Category')

# Display the results
print("Chi-Square Test of Homogeneity Results:")
print(f"Chi-Square Statistic: {results['Chi-Square Statistic']:.4f}")
print(f"P-Value: {results['P-Value']:.4f}")
print(f"Degrees of Freedom: {results['Degrees of Freedom']}")
print("\nExpected Frequencies Table:")
print(results['Expected Frequencies'])

```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

ss42_ss['Date'] = pd.to_datetime(ss42_ss['Date'], errors='coerce', dayfirst=True) # or dayfirst=False as per your date format
ss42_ss['DaysSinceReference'] = (ss42_ss['Date'] - pd.Timestamp('2000-01-01')).dt.days
data = ss42_ss[['Quantity', 'Value', 'Customs_Code', 'Weight', 'Invoice_Number', 'DaysSinceReference']]

# Test of Normality

# 1. Shapiro-Wilk test
shapiro_stat, shapiro_p = stats.shapiro(data)
print(f"Shapiro-Wilk Test: Statistic = {shapiro_stat}, p-value = {shapiro_p}")

# 2. Kolmogorov-Smirnov test
ks_stat, ks_p = stats.kstest(data, 'norm')
print(f"Kolmogorov-Smirnov Test: Statistic = {ks_stat}, p-value = {ks_p}")

# 3. Anderson-Darling test
for column in data.columns:
    anderson_result = stats.anderson(data[column], dist='norm')
    print(f"Anderson-Darling Test for {column}: Statistic = {anderson_result.statistic}")
    for i in range(len(anderson_result.critical_values)):
        print(f"  Critical value for {anderson_result.significance_level[i]}%: {anderson_result.critical_values[i]}")

# 4. Jarque-Bera test
jb_stat, jb_p = stats.jarque_bera(data)
print(f"Jarque-Bera Test: Statistic = {jb_stat}, p-value = {jb_p}")
```
